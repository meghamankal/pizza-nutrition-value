# MEAN VALUES OF EACH NUMERICAL COLUMN

```julia
using Statistics
numericcols=select(pizzas, Not([:brand,:id])) 
meanvalues=colwise(mean, numericcols) 
```

```julia
using PyPlot
using StatsPlots
pyplot()
StatsPlots.bar(names(numericcols),meanvalues,color="#EB9CB0",labels="Mean")
```

# Fat content of different brands

```julia
using Statistics
groups=groupby(pizzas,[:"brand"])  
fat_mean=[]                    
for i in 1:10                  
    x=mean(groups[i][:fat])     
    push!(fat_mean, x)        
end
```

```julia
using PyPlot
using StatsPlots
pyplot()
StatsPlots.bar(levels(pizzas["brand"]),fat_mean,color="#F08080",labels="fat value",xlabel="Brands",ylabel="fat content") #mean of each numeric columns of the data
```

# BUILDING A CALORIE CHART

```julia
groups=groupby(pizzas,[:"brand"])  
groups[1]
```

```julia
cal_mean=[]
for i in 1:10
    x=mean(groups[i][:cal])
    push!(cal_mean, x)
end

key=maximum(cal_mean)
for i in 1:length(cal_mean)
    if cal_mean[i]==key
        println(i)
        break
    end
end
```

```julia
using Plots
using PyPlot
using StatsPlots
colors=["#4BB9D1","#717D7E","#F08080","#D6EAF8","#ffcc99","#EB9CB0","#D7DBDD","#5D8FE3","#CA76DF","#A1DF76"]

title("Calorie Chart")
PyPlot.pie(cal_mean, labels=levels(pizzas["brand"]),autopct="%1.2f%%", shadow=true, explode=[0.15,0,0,0,0,0,0,0,0,0] ,colors=colors,startangle=90)
show()

```

# Building a protein chart

```julia
prot_mean=[]
for i in 1:10
    x=mean(groups[i][:prot])
    push!(prot_mean, x)
end

key=minimum(prot_mean)
for i in 1:length(prot_mean)
    if prot_mean[i]==key
        println(i)
        break
    end
end
```

```julia
using Plots
using PyPlot
using StatsPlots

pyplot()
colors=["#4BB9D1","#717D7E","#F08080","#D6EAF8","#ffcc99","#EB9CB0","#D7DBDD","#5D8FE3","#CA76DF","#A1DF76"]
title("Protein Chart")
PyPlot.pie(prot_mean, labels=levels(pizzas["brand"]),autopct="%1.2f%%",shadow=true, explode=[0,0,0,0,0.15,0,0,0,0,0] ,colors=colors, startangle=90)

show()
```
